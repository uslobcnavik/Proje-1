```python
import tkinter as tk
from tkinter import ttk, messagebox

class Urun:
    def __init__(self, ad, miktar):
        self.ad = ad
        self.miktar = miktar

    def stok_guncelle(self, miktar):
        self.miktar += miktar
        return self.miktar

class Siparis:
    def __init__(self, urun_ad, miktar):
        self.urun_ad = urun_ad
        self.miktar = miktar

class StokTakipApp:
    def __init__(self, master):
        self.master = master
        master.title("Stok Takip Sistemi")
        master.configure(bg='#f2c957')

        self.style = ttk.Style()
        self.style.theme_use('clam')
        self.style.configure('TButton', background='#4CAF50', foreground='white', font=('Arial', 10))
        self.style.configure('TLabel', background='#f0f0f0', font=('Arial', 10))
        self.style.configure('TEntry', foreground='black', font=('Arial', 10))

        self.urun_listesi = tk.Listbox(master, height=8, width=50)
        self.urun_listesi.grid(row=0, column=0, columnspan=3, padx=10, pady=10)
        self.urun_listesi.bind('<<ListboxSelect>>', self.urun_detay_goster)

        tk.Label(master, text="Ürün Adı:", bg='#f2c957').grid(row=1, column=0)
        self.urun_adi_gir = tk.Entry(master)
        self.urun_adi_gir.grid(row=1, column=1)

        tk.Label(master, text="Stok Miktarı:", bg='#f2c957').grid(row=2, column=0)
        self.urun_stok_gir = tk.Entry(master)
        self.urun_stok_gir.grid(row=2, column=1)

        self.urun_ekle_btn = ttk.Button(master, text="Ürün Ekle", command=self.urun_ekle)
        self.urun_ekle_btn.grid(row=3, column=0, columnspan=2)

        self.siparisler = [
            Siparis("Elma", 10),
            Siparis("Muz", 5)
        ]
        self.siparis_kabul_btn = ttk.Button(master, text="Siparişleri İşle", command=self.siparis_isle)
        self.siparis_kabul_btn.grid(row=4, column=0, columnspan=2)

        self.urunler = [
            Urun("Elma", 100),
            Urun("Muz", 150),
            Urun("Portakal", 200)
        ]
        for urun in self.urunler:
            self.urun_listesi.insert(tk.END, urun.ad)

    def urun_ekle(self):
        ad = self.urun_adi_gir.get()
        try:
            miktar = int(self.urun_stok_gir.get())
            yeni_urun = Urun(ad, miktar)
            self.urunler.append(yeni_urun)
            self.urun_listesi.insert(tk.END, ad)
            messagebox.showinfo("Başarılı", f"{ad} eklendi!")
        except ValueError:
            messagebox.showerror("Hata", "Geçersiz miktar girdiniz.")

    def urun_detay_goster(self, event):
        selection = self.urun_listesi.curselection()
        if selection:
            index = selection[0]
            urun = self.urunler[index]
            messagebox.showinfo("Stok Bilgisi", f"{urun.ad} - Mevcut Stok: {urun.miktar}")

    def siparis_isle(self):
        for siparis in self.siparisler:
            for urun in self.urunler:
                if urun.ad == siparis.urun_ad:
                    if urun.miktar >= siparis.miktar:
                        urun.stok_guncelle(-siparis.miktar)
                        messagebox.showinfo("Sipariş İşlendi", f"{siparis.urun_ad} için {siparis.miktar} adet sipariş işlendi.")
                    else:
                        messagebox.showwarning("Stok Yetersiz", f"{siparis.urun_ad} için yeterli stok yok.")
                    break

def main():
    root = tk.Tk()
    app = StokTakipApp(root)
    root.mainloop()

if __name__ == "__main__":
    main()
```
