```python
import tkinter as tk
from tkinter import ttk, messagebox, simpledialog

class Oyun:
    def __init__(self, ad, tur, platform):
        self.ad = ad
        self.tur = tur
        self.platform = platform
        self.degerlendirmeler = []

class Koleksiyon:
    def __init__(self):
        self.oyunlar = []

    def oyun_ekle(self, oyun):
        self.oyunlar.append(oyun)
        messagebox.showinfo("Başarılı", f"{oyun.ad} koleksiyona eklendi.")

    def oyun_bul(self, ad):
        for oyun in self.oyunlar:
            if oyun.ad == ad:
                return oyun
        return None

class Oyuncu:
    def __init__(self, ad):
        self.ad = ad
        self.favori_oyunlar = []

    def oyun_degerlendir(self, oyun, puan):
        oyun.degerlendirmeler.append((self.ad, puan))
        messagebox.showinfo("Değerlendirme", f"{oyun.ad} için {puan} puan verildi.")

class OyunKoleksiyonuApp(tk.Tk):
    def __init__(self):
        super().__init__()
        self.title('Video Oyun Koleksiyonu Yönetimi')
        self.geometry('800x600')
        self.style = ttk.Style(self)
        self.style.configure('TButton', font=('Helvetica', 12), background='#005f73', foreground='white')
        self.style.configure('TLabel', font=('Helvetica', 14), background='#333', foreground='white')
        self.style.configure('TEntry', font=('Helvetica', 12), background='#e0e0e0')
        self.configure(bg='#333')

        self.koleksiyon = Koleksiyon()
        self.oyuncular = {'Default': Oyuncu('Default')}
        self.oneri_oyunlarini_ekle()
        self.arayuz_olustur()

    def oneri_oyunlarini_ekle(self):
        oyun_listesi = [
            ("GTA V", "Aksiyon-Macera", "Çoklu Platform"),
            ("PUBG: Battlegrounds", "Battle Royale", "Çoklu Platform"),
            ("Red Dead Redemption 2", "Aksiyon-Macera", "Çoklu Platform"),
            ("Call of Duty: Warzone", "FPS", "Çoklu Platform"),
            ("FC24", "FPS", "PC"),
            ("Overwatch", "FPS", "Çoklu Platform"),
            ("Rainbow Six Siege", "Taktiksel Shooter", "Çoklu Platform"),
            ("Counter Strike 2", "FPS", "PC"),
            ("Cyberpunk 2077", "RPG", "Çoklu Platform"),
            ("Detroit: Become Human", "Macera", "Çoklu Platform")
        ]
        for oyun in oyun_listesi:
            self.koleksiyon.oyun_ekle(Oyun(*oyun))

    def arayuz_olustur(self):
        
        sekme_kontrol = ttk.Notebook(self)
        sekme_kontrol.pack(expand=1, fill='both')

        oyun_ekle_sekmesi = ttk.Frame(sekme_kontrol)
        degerlendirme_sekmesi = ttk.Frame(sekme_kontrol)
        
        sekme_kontrol.add(oyun_ekle_sekmesi, text='Oyun Ekle')
        sekme_kontrol.add(degerlendirme_sekmesi, text='Oyun Değerlendir')
       
        ttk.Label(oyun_ekle_sekmesi, text='Oyun Adı:').pack(pady=10)
        oyun_adi_girdisi = ttk.Entry(oyun_ekle_sekmesi)
        oyun_adi_girdisi.pack(pady=10)
        ttk.Label(oyun_ekle_sekmesi, text='Tür:').pack(pady=10)
        tur_girdisi = ttk.Entry(oyun_ekle_sekmesi)
        tur_girdisi.pack(pady=10)
        ttk.Label(oyun_ekle_sekmesi, text='Platform:').pack(pady=10)
        platform_girdisi = ttk.Entry(oyun_ekle_sekmesi)
        platform_girdisi.pack(pady=10)
        ttk.Button(oyun_ekle_sekmesi, text='Oyun Ekle', command=lambda: self.koleksiyon.oyun_ekle(Oyun(oyun_adi_girdisi.get(), tur_girdisi.get(), platform_girdisi.get()))).pack(pady=20)

        
        for oyun in self.koleksiyon.oyunlar:
            ttk.Button(degerlendirme_sekmesi, text=f'Değerlendir: {oyun.ad}', command=lambda o=oyun: self.degerlendirme_penceresi_olustur(o)).pack(pady=5)

    def degerlendirme_penceresi_olustur(self, oyun):
        puan = simpledialog.askinteger("Değerlendirme", f"{oyun.ad} için 1-5 arası bir puan verin", minvalue=1, maxvalue=5)
        if puan:
            self.oyuncular['Default'].oyun_degerlendir(oyun, puan)

if __name__ == "__main__":
    app = OyunKoleksiyonuApp()
    app.mainloop()
```
