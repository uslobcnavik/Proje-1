```python
import tkinter as tk
from tkinter import ttk, messagebox, scrolledtext, filedialog, simpledialog

class Ders:
    def __init__(self, ad, icerik, ogretmen):
        self.ad = ad
        self.icerik = icerik
        self.ogretmen = ogretmen
        self.materyaller = []
        self.sorular = []

    def materyal_ekle(self, materyal):
        self.materyaller.append(materyal)

    def soru_ekle(self, soru):
        self.sorular.append(soru)

class Materyal:
    def __init__(self, ad, tur, icerik):
        self.ad = ad
        self.tur = tur
        self.icerik = icerik

class Ogrenci:
    def __init__(self, isim):
        self.isim = isim
        self.dersler = []

    def ders_ekle(self, ders):
        self.dersler.append(ders)

class EgitimPlatformu(tk.Tk):
    def __init__(self):
        super().__init__()
        self.title('Eğitim Materyali Paylaşım Platformu')
        self.geometry('800x600')
        self.dersler = self.dersleri_yukle()
        self.setup_ui()

    def setup_ui(self):
        self.tab_control = ttk.Notebook(self)
        self.ders_tab = ttk.Frame(self.tab_control)
        self.materyal_tab = ttk.Frame(self.tab_control)
        self.soru_tab = ttk.Frame(self.tab_control)
        self.tab_control.add(self.ders_tab, text='Dersler')
        self.tab_control.add(self.materyal_tab, text='Materyaller')
        self.tab_control.add(self.soru_tab, text='Sorular')
        self.tab_control.pack(expand=1, fill='both')
        self.ders_listesi = tk.Listbox(self.ders_tab, height=15, width=50)
        self.ders_listesi.pack(pady=20)
        for ders in self.dersler:
            self.ders_listesi.insert(tk.END, ders.ad)
        self.materyal_listesi = tk.Listbox(self.materyal_tab, height=15, width=50)
        self.materyal_listesi.pack(pady=20)
        self.materyalleri_yukle()
        self.soru_listesi = tk.Listbox(self.soru_tab, height=15, width=50)
        self.soru_listesi.pack(pady=20)
        self.soru_ekle_butonu = ttk.Button(self.soru_tab, text="Soru Ekle", command=self.soru_ekle)
        self.soru_ekle_butonu.pack(pady=10)
    def dersleri_yukle(self):
        return [
            Ders("Tarih", "Tarih konuları ve içerikleri", "Ahmet Hoca"),
            Ders("Matematik", "Matematiksel işlemler ve problemler", "Ayşe Hoca"),
            Ders("Kimya", "Kimyasal reaksiyonlar ve deneyler", "Ali Hoca"),
            Ders("Fizik", "Fiziksel yasalar ve uygulamalar", "Mehmet Hoca"),
            Ders("Bilgisayara Giriş", "Temel bilgisayar bilimleri ve programlama", "Elif Hoca"),
            Ders("Geometri", "Geometrik şekiller ve teoremler", "Burak Hoca"),
            Ders("Biyoloji", "Canlı bilimi ve ekosistemler", "Derya Hoca")
        ]
    def materyalleri_yukle(self):
        ornek_materyaller = [
            Materyal("Algebra Notları", "PDF", "Detaylı algebra ders notları"),
            Materyal("Organik Kimya", "Video", "Organik kimya reaksiyonları videosu"),
            Materyal("Dinamik", "PDF", "Fizik dinamik konuları anlatımı"),
            Materyal("Programlama Temelleri", "Video", "Python ile programlama öğrenimi"),
            Materyal("Ekoloji", "PDF", "Ekosistemler ve çevre bilimi")
        ]
        
        for materyal in ornek_materyaller:
            self.materyal_listesi.insert(tk.END, f"{materyal.ad} ({materyal.tur})")
            self.dersler[0].materyal_ekle(materyal)
    def soru_ekle(self):
        secim = self.ders_listesi.curselection()
        if not secim:
            messagebox.showerror("Hata", "Lütfen bir ders seçiniz.")
            return
        secili_ders = self.ders_listesi.get(secim[0])
        soru = simpledialog.askstring("Soru Ekle", "Soru giriniz:")
        if not soru:
            messagebox.showwarning("Uyarı", "Boş soru girilemez.")
            return
        for ders in self.dersler:
            if ders.ad == secili_ders:
                ders.soru_ekle(soru)
                self.soru_listesi.insert(tk.END, f"{secili_ders}: {soru}")
                messagebox.showinfo("Başarılı", "Sorunuz eklendi.")
                break
if __name__ == "__main__":
    app = EgitimPlatformu()
    app.mainloop()
```
