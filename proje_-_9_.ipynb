```python
import tkinter as tk
from tkinter import ttk, messagebox

class Konaklama:
    def __init__(self, ad, fiyat):
        self.ad = ad
        self.fiyat = fiyat

class Rota:
    def __init__(self, ad, detaylar):
        self.ad = ad
        self.detaylar = detaylar
        self.degerlendirme = []

    def rota_degerlendir(self, puan):
        self.degerlendirme.append(puan)
        return f"{self.ad} rotası için ortalama puan: {sum(self.degerlendirme)/len(self.degerlendirme):.2f}"

class Seyahat:
    def __init__(self):
        self.rotalar = [
            Rota("Great Ocean Road, Avustralya", "Okyanus boyunca uzanan muhteşem manzaralı yol."),
            Rota("California State Route 1, ABD", "Pacifik sahil şeridi boyunca ikonik bir Amerikan yolculuğu."),
            Rota("Ring Road, İzlanda", "İzlanda'nın doğal güzelliklerini dolaşan tam bir tur."),
            Rota("Toskana, İtalya", "Toskana'nın şarap ülkesi boyunca güzel köyleri ve tepeleri."),
            Rota("South Island, Yeni Zelanda", "Yeni Zelanda'nın güney adasının etkileyici manzaraları.")
        ]
        self.konaklamalar = []

    def plan_duzenle(self):
        plan = "Seyahat Planı:\n"
        for rota in self.rotalar:
            plan += f"Rota: {rota.detaylar}\n"
        for konaklama in self.konaklamalar:
            plan += f"Konaklama: {konaklama.ad}, Fiyat: {konaklama.fiyat} $\n"
        return plan

class SeyahatUygulamasi(tk.Tk):
    def __init__(self):
        super().__init__()
        self.title('Seyahat Planlama Uygulaması')
        self.geometry('300x350')
        self.configure(bg='#34495e')  
        self.seyahat = Seyahat()
        self.arayuz_olustur()

    def arayuz_olustur(self):
        self.stil = ttk.Style()
        self.stil.configure('TButton', background='#16a085', foreground='white', font=('Arial', 12, 'bold'))
        self.stil.configure('TLabel', background='#34495e', foreground='white', font=('Arial', 12))
        self.stil.configure('TEntry', foreground='black', font=('Arial', 12))
 
        self.rota_listesi = tk.Listbox(self, height=5, font=('Arial', 12), bg='#ecf0f1', fg='#2c3e50')
        self.rota_listesi.grid(row=0, column=0, columnspan=2, padx=10, pady=10)
        for rota in self.seyahat.rotalar:
            self.rota_listesi.insert(tk.END, rota.ad)

        self.plan_butonu = ttk.Button(self, text='Planı Görüntüle', command=self.plan_goster)
        self.plan_butonu.grid(row=1, column=0, columnspan=2, padx=10, pady=10)

    def plan_goster(self):
        plan = self.seyahat.plan_duzenle()
        messagebox.showinfo("Seyahat Planı", plan)

if __name__ == "__main__":
    uygulama = SeyahatUygulamasi()
    uygulama.mainloop()
```

```python
import tkinter as tk
from tkinter import ttk, messagebox, simpledialog

class Konaklama:
    def __init__(self, ad, fiyat):
        self.ad = ad
        self.fiyat = fiyat

class Rota:
    def __init__(self, ad, detaylar):
        self.ad = ad
        self.detaylar = detaylar
        self.degerlendirme = []

    def rota_degerlendir(self, puan):
        self.degerlendirme.append(puan)
        return f"{self.ad} rotası için ortalama puan: {sum(self.degerlendirme)/len(self.degerlendirme):.2f}"

class Seyahat:
    def __init__(self):
        self.rotalar = []
        self.konaklamalar = []

    def rota_ekle(self, rota):
        self.rotalar.append(rota)

    def konaklama_ekle(self, konaklama):
        self.konaklamalar.append(konaklama)

    def plan_duzenle(self):
        plan = "Seyahat Planı:\n"
        for rota in self.rotalar:
            plan += f"Rota: {rota.detaylar}\n"
        for konaklama in self.konaklamalar:
            plan += f"Konaklama: {konaklama.ad}, Fiyat: {konaklama.fiyat} TL\n"
        return plan

class SeyahatUygulamasi(tk.Tk):
    def __init__(self):
        super().__init__()
        self.title('Seyahat Planlama Uygulaması')
        self.geometry('400x600')
        self.configure(bg='#f0f0f0')
        self.seyahat = Seyahat()
        self.arayuz_olustur()

    def arayuz_olustur(self):
        self.stil = ttk.Style()
        self.stil.configure('TButton', background='#333', foreground='white', font=('Arial', 10))
        self.stil.configure('TLabel', background='#f0f0f0', font=('Arial', 10))
        self.stil.configure('TEntry', foreground='black', font=('Arial', 10))

        ttk.Label(self, text='Rota Adı:').grid(row=0, column=0, padx=10, pady=10)
        self.rota_ad_gir = ttk.Entry(self)
        self.rota_ad_gir.grid(row=0, column=1, padx=10, pady=10)

        ttk.Label(self, text='Rota Detayları:').grid(row=1, column=0, padx=10, pady=10)
        self.rota_detay_gir = ttk.Entry(self)
        self.rota_detay_gir.grid(row=1, column=1, padx=10, pady=10)

        self.rota_ekle_btn = ttk.Button(self, text='Rota Ekle', command=self.rota_ekle)
        self.rota_ekle_btn.grid(row=2, column=1, padx=10, pady=10)

        ttk.Label(self, text='Konaklama Adı:').grid(row=3, column=0, padx=10, pady=10)
        self.konaklama_ad_gir = ttk.Entry(self)
        self.konaklama_ad_gir.grid(row=3, column=1, padx=10, pady=10)

        ttk.Label(self, text='Fiyat:').grid(row=4, column=0, padx=10, pady=10)
        self.konaklama_fiyat_gir = ttk.Entry(self)
        self.konaklama_fiyat_gir.grid(row=4, column=1, padx=10, pady=10)

        self.konaklama_ekle_btn = ttk.Button(self, text='Konaklama Ekle', command=self.konaklama_ekle)
        self.konaklama_ekle_btn.grid(row=5, column=1, padx=10, pady=10)

        self.plan_buton = ttk.Button(self, text='Planı Görüntüle', command=self.plan_goster)
        self.plan_buton.grid(row=6, column=1, padx=10, pady=20)

    def rota_ekle(self):
        ad = self.rota_ad_gir.get()
        detaylar = self.rota_detay_gir.get()
        rota = Rota(ad, detaylar)
        self.seyahat.rota_ekle(rota)
        messagebox.showinfo("Başarılı", f"Rota '{rota.ad}' eklendi")
        self.rota_ad_gir.delete(0, tk.END)
        self.rota_detay_gir.delete(0, tk.END)

    def konaklama_ekle(self):
        ad = self.konaklama_ad_gir.get()
        fiyat = float(self.konaklama_fiyat_gir.get())
        konaklama = Konaklama(ad, fiyat)
        self.seyahat.konaklama_ekle(konaklama)
        messagebox.showinfo("Başarılı", f"Konaklama '{konaklama.ad}' eklendi")
        self.konaklama_ad_gir.delete(0, tk.END)
        self.konaklama_fiyat_gir.delete(0, tk.END)

    def plan_goster(self):
        plan = self.seyahat.plan_duzenle()
        messagebox.showinfo("Seyahat Planı", plan)

if __name__ == "__main__":
    uygulama = SeyahatUygulamasi()
    uygulama.mainloop()
```
