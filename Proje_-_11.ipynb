```python
import tkinter as tk
from tkinter import ttk, messagebox

class Enstruman:
    def __init__(self, ad, stok):
        self.ad = ad
        self.stok = stok

    def stok_guncelle(self, miktar):
        self.stok += miktar
        return self.stok

class Musteri:
    def __init__(self, ad, siparisler=None):
        self.ad = ad
        if siparisler is None:
            self.siparisler = []

    def siparis_ekle(self, siparis):
        self.siparisler.append(siparis)

class Siparis:
    def __init__(self, siparis_numarasi, enstruman, miktar):
        self.siparis_numarasi = siparis_numarasi
        self.enstruman = enstruman
        self.miktar = miktar

class DestekTalebi:
    def __init__(self, talep_numarasi, detaylar):
        self.talep_numarasi = talep_numarasi
        self.detaylar = detaylar

class MuzikDukkanApp(tk.Tk):
    def __init__(self):
        super().__init__()
        self.title('Müzik Enstrümanı Dükkanı Yönetimi')
        self.geometry('600x500')
        self.configure(bg='#967ca3')
        self.arayuz_ayarla()

    def arayuz_ayarla(self):
        self.stil = ttk.Style()
        self.stil.theme_use('clam')
        self.stil.configure('TButton', background='#3a154d', foreground='white', font=('Arial', 12, 'bold'))
        self.stil.configure('TLabel', background='#f4f2f5', font=('Arial', 12))
        self.stil.configure('TEntry', font=('Arial', 12), width=20)

        ttk.Label(self, text='Enstrüman Adı:').grid(row=0, column=0, padx=10, pady=10)
        self.enstruman_ad_girisi = ttk.Entry(self)
        self.enstruman_ad_girisi.grid(row=0, column=1, padx=10, pady=10)

        ttk.Label(self, text='Stok Miktarı:').grid(row=1, column=0, padx=10, pady=10)
        self.stok_miktari_girisi = ttk.Entry(self)
        self.stok_miktari_girisi.grid(row=1, column=1, padx=10, pady=10)

        self.stok_ekle_butonu = ttk.Button(self, text='Stok Ekle', command=self.stok_ekle)
        self.stok_ekle_butonu.grid(row=0, column=2, padx=10, pady=10)

        ttk.Label(self, text='Müşteri Adı:').grid(row=2, column=0, padx=10, pady=10)
        self.musteri_adi_girisi = ttk.Entry(self)
        self.musteri_adi_girisi.grid(row=2, column=1, padx=10, pady=10)

        self.musteri_ekle_butonu = ttk.Button(self, text='Müşteri Ekle', command=self.musteri_ekle)
        self.musteri_ekle_butonu.grid(row=2, column=2, padx=10, pady=10)

        ttk.Label(self, text='Sipariş Numarası:').grid(row=3, column=0, padx=10, pady=10)
        self.siparis_numarasi_girisi = ttk.Entry(self)
        self.siparis_numarasi_girisi.grid(row=3, column=1, padx=10, pady=10)

        self.siparis_isle_butonu = ttk.Button(self, text='Sipariş İşle', command=self.siparis_isle)
        self.siparis_isle_butonu.grid(row=3, column=2, padx=10, pady=10)

    def stok_ekle(self):
        
        messagebox.showinfo("Başarılı", "Stok başarıyla eklendi")

    def musteri_ekle(self):
       
        messagebox.showinfo("Başarılı", "Müşteri başarıyla eklendi")

    def siparis_isle(self):
        
        messagebox.showinfo("Başarılı", "Sipariş başarıyla işlendi")

if __name__ == "__main__":
    uygulama = MuzikDukkanApp()
    uygulama.mainloop()
```
