```python
import tkinter as tk
from tkinter import ttk, messagebox, simpledialog

class Sporcu:
    def __init__(self, isim, spor_dali):
        self.isim = isim
        self.spor_dali = spor_dali
        self.antrenmanlar = []

    def program_olustur(self, ad, tarih, detaylar):
        yeni_antrenman = Antrenman(ad, tarih, detaylar)
        self.antrenmanlar.append(yeni_antrenman)
        return f"{ad} adlı antrenman {self.isim}'in programına eklendi."

    def rapor_al(self):
        rapor = f"{self.isim} için Antrenman Raporu:\n"
        for antrenman in self.antrenmanlar:
            rapor += f"{antrenman.ad} - {antrenman.tarih}: {antrenman.detaylar}\n"
        return rapor

class Antrenman:
    def __init__(self, ad, tarih, detaylar):
        self.ad = ad
        self.tarih = tarih
        self.detaylar = detaylar

class SporTakipUygulamasi(tk.Tk):
    def __init__(self):
        super().__init__()
        self.title("Spor Takip Uygulaması")

        self.sporcular = [
            Sporcu("Şeyda Yahşi", "Yüzme"),
            Sporcu("Kerem Ak", "Atletizm"),
            Sporcu("Gökhan Sine", "Jimnastik"),
            Sporcu("Doğukan Çor", "Futbol")
        ]
        self.sporcular[0].program_olustur("Pazartesi Yüzme", "2024-01-01", "50m serbest stil")
        self.sporcular[1].program_olustur("Salı Koşu", "2024-01-02", "10km koşu")
        self.sporcular[2].program_olustur("Perşembe Jimnastik", "2024-01-04", "Barfiks çalışması")
        self.sporcular[3].program_olustur("Cuma Futbol", "2024-01-05", "Dribbling ve şut çalışması")
        
        self.sp_liste = tk.Listbox(self, height=4)
        self.sp_liste.pack(padx=5, pady=5, fill=tk.BOTH, expand=True)
        self.sp_liste.bind('<<ListboxSelect>>', self.detaylari_goster)
        for sp in self.sporcular:
            self.sp_liste.insert(tk.END, sp.isim)

        self.detaylar_metin = tk.Text(self, height=6)
        self.detaylar_metin.pack(padx=5, pady=5, fill=tk.BOTH, expand=True)
        
        
        tk.Label(self, text="Antrenman Adı:").pack()
        self.antrenman_ad_gir = tk.Entry(self)
        self.antrenman_ad_gir.pack(fill=tk.X, padx=5, pady=2)

        tk.Label(self, text="Tarih:").pack()
        self.tarih_gir = tk.Entry(self)
        self.tarih_gir.pack(fill=tk.X, padx=5, pady=2)

        tk.Label(self, text="Detaylar:").pack()
        self.detaylar_gir = tk.Entry(self)
        self.detaylar_gir.pack(fill=tk.X, padx=5, pady=2)

        self.ekle_buton = tk.Button(self, text="Antrenman Ekle", command=self.antrenman_ekle)
        self.ekle_buton.pack(pady=5)

    def detaylari_goster(self, event=None):
        secim = self.sp_liste.curselection()
        if secim:
            index = secim[0]
            sp = self.sporcular[index]
            rapor = sp.rapor_al()
            self.detaylar_metin.delete('1.0', tk.END)
            self.detaylar_metin.insert('1.0', rapor)

    def antrenman_ekle(self):
        index = self.sp_liste.curselection()[0]
        sp = self.sporcular[index]
        ad = self.antrenman_ad_gir.get()
        tarih = self.tarih_gir.get()
        detaylar = self.detaylar_gir.get()
        mesaj = sp.program_olustur(ad, tarih, detaylar)
        self.detaylar_metin.insert(tk.END, f"\n{mesaj}")
        self.antrenman_ad_gir.delete(0, tk.END)
        self.tarih_gir.delete(0, tk.END)
        self.detaylar_gir.delete(0, tk.END)
        
def ana_program():
    root = tk.Tk()
    app = SporTakipUygulamasi()
    root.mainloop()
if __name__ == "__main__":
    ana_program()
```
