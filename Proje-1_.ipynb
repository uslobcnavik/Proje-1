```python
import tkinter as tk
from tkinter import ttk, messagebox

class Arac:
    def __init__(self, arac_id, model, durum="müsait"):
        self.arac_id = arac_id
        self.model = model
        self.durum = durum  # müsait, kiralandi

    def arac_durum_guncelle(self, yeni_durum):
        self.durum = yeni_durum

class Musteri:
    def __init__(self, musteri_id, isim, telefon):
        self.musteri_id = musteri_id
        self.isim = isim
        self.telefon = telefon

class Kiralama:
    def __init__(self, musteri, arac):
        self.musteri = musteri
        self.arac = arac

    def kiralama_yap(self):
        if self.arac.durum == "müsait":
            self.arac.arac_durum_guncelle("kiralandi")
            return "Kiralama başarılı!"
        else:
            return "Araç müsait değil."

    def kiralama_iptal_et(self):
        if self.arac.durum == "kiralandi":
            self.arac.arac_durum_guncelle("müsait")
            return "Kiralama iptal edildi."
        else:
            return "Kiralama iptal edilemez, araç zaten müsait."

class AracKiralamaGUI:
    def __init__(self, root):
        self.root = root
        root.title("Araç Kiralama Sistemi")
        root.configure(bg='light gray')

        self.araclar = [
            Arac(1, "Toyota Corolla"),
            Arac(2, "Honda Civic"),
            Arac(3, "Volvo XC90"),
            Arac(4, "BMW 3 Series")
        ]

        self.musteriler = [
            Musteri(1, "Ozan Abes", "555-1234"),
            Musteri(2, "Umutcan Cevik", "555-5678"),
            Musteri(3, "Zekeriya Kurt", "555-1238"),
            Musteri(4, "Kıvanc Bolsu", "555-9101")
        ]

        style = ttk.Style()
        style.configure('TLabel', background='light gray', font=('Arial', 10))
        style.configure('TButton', background='light gray', font=('Arial', 10))
        style.configure('TCombobox', background='white', font=('Arial', 10))

        self.arac_combobox = ttk.Combobox(root, values=[arac.model for arac in self.araclar], state="readonly")
        self.arac_combobox.grid(row=0, column=1, padx=10, pady=10)
        ttk.Label(root, text="Araç Seçiniz:", background='light gray').grid(row=0, column=0)

        self.musteri_combobox = ttk.Combobox(root, values=[musteri.isim for musteri in self.musteriler], state="readonly")
        self.musteri_combobox.grid(row=1, column=1, padx=10, pady=10)
        ttk.Label(root, text="Müşteri Seçiniz:", background='light gray').grid(row=1, column=0)

        self.kirala_button = tk.Button(root, text="Araç Kirala", command=self.kiralama_yap, bg='green', fg='white', font=('Arial', 10))
        self.kirala_button.grid(row=2, column=0, padx=10, pady=10, columnspan=2)

        self.iptal_button = tk.Button(root, text="Kiralama İptal Et", command=self.kiralama_iptal, bg='red', fg='white', font=('Arial', 10))
        self.iptal_button.grid(row=3, column=0, padx=10, pady=10, columnspan=2)

    def kiralama_yap(self):
        arac_index = self.arac_combobox.current()
        musteri_index = self.musteri_combobox.current()

        if arac_index == -1 or musteri_index == -1:
            messagebox.showerror("Hata", "Lütfen araç ve müşteri seçimi yapınız!")
            return

        secilen_arac = self.araclar[arac_index]
        secilen_musteri = self.musteriler[musteri_index]
        kiralama = Kiralama(secilen_musteri, secilen_arac)

        sonuc = kiralama.kiralama_yap()
        messagebox.showinfo("Kiralama Sonucu", sonuc)

    def kiralama_iptal(self):
        arac_index = self.arac_combobox.current()

        if arac_index == -1:
            messagebox.showerror("Hata", "Lütfen bir araç seçiniz!")
            return

        secilen_arac = self.araclar[arac_index]
        sonuc = Kiralama(None, secilen_arac).kiralama_iptal_et()
        messagebox.showinfo("Kiralama İptali", sonuc)

if __name__ == "__main__":
    root = tk.Tk()
    app = AracKiralamaGUI(root)
    root.mainloop()
```
