```python
import tkinter as tk
from tkinter import ttk, messagebox

class SaglikUygulamasi(tk.Tk):
    def __init__(self):
        super().__init__()
        self.title('Kişisel Sağlık Takip Uygulaması')
        self.geometry('600x300')
        self.configure(bg='#450808')
        self.arayuz_ayarla()

    def arayuz_ayarla(self):
        self.stil = ttk.Style()
        self.stil.theme_use('clam')
        self.stil.configure('TButton', background='#2f79b5', foreground='white', font=('Arial', 12, 'bold'))
        self.stil.configure('TLabel', background='#f7f2f3', font=('Arial', 12))
        self.stil.configure('TEntry', font=('Arial', 12), width=20)

        ttk.Label(self, text='Ağırlık (kg):').grid(row=0, column=0, padx=10, pady=10)
        self.agirlik_girisi = ttk.Entry(self)
        self.agirlik_girisi.grid(row=0, column=1, padx=10, pady=10)

        ttk.Label(self, text='Boy (cm):').grid(row=1, column=0, padx=10, pady=10)
        self.boy_girisi = ttk.Entry(self)
        self.boy_girisi.grid(row=1, column=1, padx=10, pady=10)

        ttk.Label(self, text='Kan Basıncı:').grid(row=2, column=0, padx=10, pady=10)
        self.kan_basinci_girisi = ttk.Entry(self)
        self.kan_basinci_girisi.grid(row=2, column=1, padx=10, pady=10)

        self.saglik_verisi_kaydet_ve_egzersiz_oner_button = ttk.Button(self, text='Sağlık Verisi Kaydet ve Egzersiz Öner', command=self.egzersiz_oner)
        self.saglik_verisi_kaydet_ve_egzersiz_oner_button.grid(row=3, column=1, padx=10, pady=10)

    def egzersiz_oner(self):
        agirlik = float(self.agirlik_girisi.get())
        boy = float(self.boy_girisi.get())
        bmi = self.vucut_kitle_indeksi_hesapla(agirlik, boy)
        egzersiz_onerileri = self.egzersiz_onerileri_al(bmi)
        mesaj = f"Önerilen Egzersizler: {egzersiz_onerileri}"
        messagebox.showinfo("Egzersiz Önerileri", mesaj)

    def vucut_kitle_indeksi_hesapla(self, agirlik, boy):
        boy_metre_cinsinden = boy / 100
        vucut_kitle_indeksi = agirlik / (boy_metre_cinsinden ** 2)
        return vucut_kitle_indeksi

    def egzersiz_onerileri_al(self, bmi):
        if bmi < 18.5:
            return "Hafif ya da orta tempoda yüzme"
        elif 18.5 <= bmi < 25:
            return "Ağırlık kaldırma, Tempolu yürüyüş, Cardiyo"
        elif 25 <= bmi < 30:
            return "Direnç Egzersizleri, Cardiyo , Tempolu yürüyüş"
        else:
            return "Direnç Egzersizleri, Cardiyo , Tempolu yürüyüş , bisiklet sürme , Nefes egzersizleri"

if __name__ == "__main__":
    uygulama = SaglikUygulamasi()
    uygulama.mainloop()
```

```python
import tkinter as tk
from tkinter import ttk, messagebox

class SaglikUygulamasi(tk.Tk):
    def __init__(self):
        super().__init__()
        self.title('Kişisel Sağlık Takip Uygulaması')
        self.geometry('600x500')
        self.configure(bg='#450808')
        self.arayuz_ayarla()

    def arayuz_ayarla(self):
        self.stil = ttk.Style()
        self.stil.theme_use('clam')
        self.stil.configure('TButton', background='#2f79b5', foreground='white', font=('Arial', 12, 'bold'))
        self.stil.configure('TLabel', background='#f7f2f3', font=('Arial', 12))
        self.stil.configure('TEntry', font=('Arial', 12), width=20)

        ttk.Label(self, text='Ağırlık (kg):').grid(row=0, column=0, padx=10, pady=10)
        self.agirlik_girisi = ttk.Entry(self)
        self.agirlik_girisi.grid(row=0, column=1, padx=10, pady=10)

        ttk.Label(self, text='Boy (cm):').grid(row=1, column=0, padx=10, pady=10)
        self.boy_girisi = ttk.Entry(self)
        self.boy_girisi.grid(row=1, column=1, padx=10, pady=10)

        ttk.Label(self, text='Kan Basıncı:').grid(row=2, column=0, padx=10, pady=10)
        self.kan_basinci_girisi = ttk.Entry(self)
        self.kan_basinci_girisi.grid(row=2, column=1, padx=10, pady=10)

        self.saglik_verisi_kaydet_button = ttk.Button(self, text='Sağlık Verisi Kaydet', command=self.saglik_verisi_kaydet)
        self.saglik_verisi_kaydet_button.grid(row=3, column=1, padx=10, pady=10)

        ttk.Label(self, text='Egzersiz Türü:').grid(row=4, column=0, padx=10, pady=10)
        self.egzersiz_turu_girisi = ttk.Entry(self)
        self.egzersiz_turu_girisi.grid(row=4, column=1, padx=10, pady=10)

        ttk.Label(self, text='Egzersiz Süresi (dakika):').grid(row=5, column=0, padx=10, pady=10)
        self.egzersiz_suresi_girisi = ttk.Entry(self)
        self.egzersiz_suresi_girisi.grid(row=5, column=1, padx=10, pady=10)

        self.egzersiz_kaydet_button = ttk.Button(self, text='Egzersiz Kaydet', command=self.egzersiz_kaydet)
        self.egzersiz_kaydet_button.grid(row=6, column=1, padx=10, pady=10)

    def saglik_verisi_kaydet(self):
        messagebox.showinfo("Başarılı", "Sağlık verileri başarıyla kaydedildi")

    def egzersiz_kaydet(self):
        messagebox.showinfo("Başarılı", "Egzersiz başarıyla kaydedildi")

if __name__ == "__main__":
    uygulama = SaglikUygulamasi()
    uygulama.mainloop()
```
