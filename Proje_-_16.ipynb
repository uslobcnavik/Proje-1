```python
import tkinter as tk
from tkinter import ttk, messagebox, scrolledtext, simpledialog

class Film:
    def __init__(self, ad, yonetmen, tur, sure):
        self.ad = ad
        self.yonetmen = yonetmen
        self.tur = tur
        self.sure = sure
        self.yorumlar = []
        self.izleme_listesi = []
        self.degerlendirmeler = []

    def yorum_ekle(self, yorum):
        self.yorumlar.append(yorum)

    def degerlendirme_ekle(self, degerlendirme):
        self.degerlendirmeler.append(degerlendirme)

class FilmIzlemeServisi(tk.Tk):
    def __init__(self):
        super().__init__()
        self.title('Film ve Dizi İzleme Servisi')
        self.geometry('800x700')
        self.configure(bg='#827979')
        self.filmler = self.filmleri_yukle()
        self.setup_ui()

    def filmleri_yukle(self):
        return [
            Film("The Godfather", "Francis Ford Coppola", "Suç, Drama", 175),
            Film("The Dark Knight", "Christopher Nolan", "Aksiyon, Suç, Drama", 152),
            Film("Forrest Gump", "Robert Zemeckis", "Drama, Romantik", 142),
            Film("Interstellar", "Christopher Nolan", "Macera, Drama, Bilim Kurgu", 169),
            Film("The Matrix", "Lana Wachowski, Lilly Wachowski", "Aksiyon, Bilim Kurgu", 136)
        ]

    def setup_ui(self):
        self.film_listesi = tk.Listbox(self, height=10, width=50)
        self.film_listesi.grid(row=0, column=0, padx=10, pady=10)
        for film in self.filmler:
            self.film_listesi.insert(tk.END, film.ad)

        self.yorum_metni = scrolledtext.ScrolledText(self, height=10, width=50)
        self.yorum_metni.grid(row=0, column=1, padx=10, pady=10)
        self.yorum_ekle_butonu = ttk.Button(self, text='Yorum Ekle', command=self.yorum_ekle)
        self.yorum_ekle_butonu.grid(row=1, column=1, padx=10, pady=10)
        
        self.degerlendirme_girisi = tk.Entry(self)
        self.degerlendirme_girisi.grid(row=2, column=1, padx=10, pady=10)
        self.degerlendirme_butonu = ttk.Button(self, text='Değerlendir', command=self.degerlendir)
        self.degerlendirme_butonu.grid(row=3, column=1, padx=10, pady=10)

        ttk.Label(self, text='Film Adı:').grid(row=4, column=0, padx=10, pady=10)
        self.yeni_film_adi = ttk.Entry(self)
        self.yeni_film_adi.grid(row=4, column=1, padx=10, pady=10)

        ttk.Label(self, text='Yönetmen:').grid(row=5, column=0, padx=10, pady=10)
        self.yeni_film_yonetmeni = ttk.Entry(self)
        self.yeni_film_yonetmeni.grid(row=5, column=1, padx=10, pady=10)

        ttk.Label(self, text='Tür:').grid(row=6, column=0, padx=10, pady=10)
        self.yeni_film_turu = ttk.Entry(self)
        self.yeni_film_turu.grid(row=6, column=1, padx=10, pady=10)

        ttk.Label(self, text='Süre (dk):').grid(row=7, column=0, padx=10, pady=10)
        self.yeni_film_suresi = ttk.Entry(self)
        self.yeni_film_suresi.grid(row=7, column=1, padx=10, pady=10)

        self.film_ekle_butonu = ttk.Button(self, text='Film Ekle', command=self.film_ekle)
        self.film_ekle_butonu.grid(row=8, column=1, padx=10, pady=10)

    def yorum_ekle(self):
        secim = self.film_listesi.curselection()
        if secim:
            index = secim[0]
            film = self.filmler[index]
            yorum = simpledialog.askstring("Yorum Yap", "Film hakkında yorumunuzu yazın:")
            if yorum:
                film.yorum_ekle(yorum)
                self.yorum_metni.insert(tk.END, yorum + '\n')
                messagebox.showinfo("Başarılı", "Yorumunuz eklendi.")

    def degerlendir(self):
        secim = self.film_listesi.curselection()
        if secim:
            index = secim[0]
            film = self.filmler[index]
            degerlendirme = self.degerlendirme_girisi.get()
            if degerlendirme:
                film.degerlendirme_ekle(degerlendirme)
                messagebox.showinfo("Başarılı", "Film değerlendirildi.")
                self.degerlendirme_girisi.delete(0, tk.END)

    def film_ekle(self):
        ad = self.yeni_film_adi.get()
        yonetmen = self.yeni_film_yonetmeni.get()
        tur = self.yeni_film_turu.get()
        sure = self.yeni_film_suresi.get()
        if ad and yonetmen and tur and sure:
            yeni_film = Film(ad, yonetmen, tur, int(sure))
            self.filmler.append(yeni_film)
            self.film_listesi.insert(tk.END, yeni_film.ad)
            messagebox.showinfo("Başarılı", f"{ad} başarıyla eklendi.")
            self.yeni_film_adi.delete(0, tk.END)
            self.yeni_film_yonetmeni.delete(0, tk.END)
            self.yeni_film_turu.delete(0, tk.END)
            self.yeni_film_suresi.delete(0, tk.END)
        else:
            messagebox.showerror("Hata", "Lütfen tüm alanları doldurun.")

if __name__ == "__main__":
    app = FilmIzlemeServisi()
    app.mainloop()
```
