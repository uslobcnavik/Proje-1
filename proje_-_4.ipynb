```python
import tkinter as tk
from tkinter import ttk, messagebox

class Etkinlik:
    def __init__(self, etkinlik_id, ad, tarih, yer):
        self.etkinlik_id = etkinlik_id
        self.ad = ad
        self.tarih = tarih
        self.yer = yer
        self.katilimcilar = []

    def katilimci_ekle(self, katilimci):
        if katilimci not in self.katilimcilar:
            self.katilimcilar.append(katilimci)
            return f"{katilimci.isim} {self.ad} etkinliğine eklendi."
        return f"{katilimci.isim} zaten {self.ad} etkinliğine katılmış."

class Katilimci:
    def __init__(self, katilimci_id, isim, email):
        self.katilimci_id = katilimci_id
        self.isim = isim
        self.email = email

class Bilet:
    def __init__(self, bilet_id, etkinlik, katilimci):
        self.bilet_id = bilet_id
        self.etkinlik = etkinlik
        self.katilimci = katilimci

    def bilet_yazdir(self):
        return f"Bilet ID: {self.bilet_id}, Etkinlik: {self.etkinlik.ad}, Katılımcı: {self.katilimci.isim}, Yer: {self.etkinlik.yer}, Tarih: {self.etkinlik.tarih}"

class EtkinlikGUI:
    def __init__(self, master):
        self.master = master
        master.title("Etkinlik Yönetim Sistemi")
        master.configure(bg='#0a3b11')

        self.etkinlikler = [
            Etkinlik("1", "Konser", "2024-06-12", "Kuruçeşme Açık Hava"),
            Etkinlik("2", "Tiyatro", "2024-05-13", "Asmalı Sahne"),
            Etkinlik("3", "Stand-Up", "2024-05-06", "Moi Sahne"),
            Etkinlik("4", "Holiday Fest", "2023-12-31", "Life Park")
        ]
        self.katilimcilar = [
            Katilimci("1", "Ali İnik", "ali@gmail.com"),
            Katilimci("2", "Sıla Polat", "sila@gmail.com"),
            Katilimci("3", "Rahmi Narinlik", "rahmi@gmail.com")
        ]
        self.biletler = []

        style = ttk.Style()
        style.configure('TLabel', background='#2fa13f', font=('Helvetica', 12))
        style.configure('TButton', background='#005f73', foreground='green', font=('Helvetica', 12))
        style.configure('TCombobox', background='green', font=('Helvetica', 12))

        self.etkinlik_secim = ttk.Combobox(master, values=[etkinlik.ad for etkinlik in self.etkinlikler], state="readonly", width=30)
        self.etkinlik_secim.grid(row=0, column=1, padx=10, pady=10)
        ttk.Label(master, text="Etkinlik seçiniz:").grid(row=0, column=0)

        self.katilimci_secim = ttk.Combobox(master, values=[katilimci.isim for katilimci in self.katilimcilar], state="readonly", width=30)
        self.katilimci_secim.grid(row=1, column=1, padx=10, pady=10)
        ttk.Label(master, text="Katılımcı seçiniz:").grid(row=1, column=0)

        ttk.Button(master, text="Katılımcı Ekle", command=self.katilimci_ekle, style='TButton').grid(row=2, column=0, columnspan=2, padx=10, pady=10)
        ttk.Button(master, text="Bilet Yazdır", command=self.bilet_yazdir, style='TButton').grid(row=3, column=0, columnspan=2, padx=10, pady=10)

    def katilimci_ekle(self):
        secilen_etkinlik = self.etkinlikler[self.etkinlik_secim.current()]
        secilen_katilimci = self.katilimcilar[self.katilimci_secim.current()]
        sonuc = secilen_etkinlik.katilimci_ekle(secilen_katilimci)
        messagebox.showinfo("Katılım İşlemi", sonuc)

    def bilet_yazdir(self):
        secilen_etkinlik = self.etkinlikler[self.etkinlik_secim.current()]
        secilen_katilimci = self.katilimcilar[self.katilimci_secim.current()]
        yeni_bilet = Bilet(len(self.biletler)+1, secilen_etkinlik, secilen_katilimci)
        self.biletler.append(yeni_bilet)
        sonuc = yeni_bilet.bilet_yazdir()
        messagebox.showinfo("Bilet Yazdırma İşlemi", sonuc)

def main():
    root = tk.Tk()
    app = EtkinlikGUI(root)
    root.mainloop()

if __name__ == "__main__":
    main()
```
