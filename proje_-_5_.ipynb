```python
import tkinter as tk
from tkinter import ttk, messagebox

class Kurs:
    def __init__(self, kurs_id, ad, egitmen, icerik):
        self.kurs_id = kurs_id
        self.ad = ad
        self.egitmen = egitmen
        self.icerik = icerik
        self.kayitli_ogrenciler = []

    def kursa_ogrenci_ekle(self, ogrenci):
        self.kayitli_ogrenciler.append(ogrenci)
        return f"{ogrenci.isim}, {self.ad} kursuna kaydoldu."

class Egitmen:
    def __init__(self, isim, uzmanlik):
        self.isim = isim
        self.uzmanlik = uzmanlik

class Ogrenci:
    def __init__(self, isim, email):
        self.isim = isim
        self.email = email

class EgitimGUI:
    def __init__(self, master):
        self.master = master
        master.title("Online Eğitim Platformu")
        master.configure(bg='#f2dec2')

        self.egitmenler = [Egitmen("Şenol Öğretmen", "Matematik"), Egitmen("Koray Öğretmen", "Fizik"), Egitmen("Ayşe Öğretmen", "Kimya")]
        self.kurslar = [Kurs("1", "Matematik 101", self.egitmenler[0], "Temel Matematik Kavramları"), 
                        Kurs("2", "Fizik 201", self.egitmenler[1], "Temel Fizik Kavramları"),
                        Kurs("3", "Kimya 301", self.egitmenler[2], "Temel Kimya"),
                        Kurs("4", "Bilgisayar 431", self.egitmenler[0], "Excel")]

        self.ogrenciler = [Ogrenci("Kemal Küçük", "kemal@gmail.com"), Ogrenci("Alice Swan", "alicewan@gmail.com"),
                          Ogrenci("Mehmet Yılmaz", "mehmet@gmail.com"), Ogrenci("Ayşe Demir", "ayse@gmail.com"),
                          Ogrenci("Ahmet Yıldırım", "ahmet@gmail.com")]

        style = ttk.Style()
        style.configure('TLabel', background='#d6ccbf', font=('Helvetica', 12))
        style.configure('TButton', background='#333333', foreground='black', font=('Helvetica', 12), relief='raised')
        style.configure('TCombobox', background='black', font=('Helvetica', 12))

        self.kurs_secim = ttk.Combobox(master, values=[kurs.ad for kurs in self.kurslar], state="readonly", width=30)
        self.kurs_secim.grid(row=0, column=1, padx=10, pady=10)
        ttk.Label(master, text="Kurs seçiniz:").grid(row=0, column=0)

        self.ogrenci_secim = ttk.Combobox(master, values=[ogrenci.isim for ogrenci in self.ogrenciler], state="readonly", width=30)
        self.ogrenci_secim.grid(row=1, column=1, padx=10, pady=10)
        ttk.Label(master, text="Öğrenci seçiniz:").grid(row=1, column=0)

        ttk.Button(master, text="Kursa Kaydol", command=self.kursa_kaydol, style='TButton').grid(row=2, column=0, columnspan=2, padx=10, pady=10)

    def kursa_kaydol(self):
        secilen_kurs = self.kurslar[self.kurs_secim.current()]
        secilen_ogrenci = self.ogrenciler[self.ogrenci_secim.current()]
        sonuc = secilen_kurs.kursa_ogrenci_ekle(secilen_ogrenci)
        messagebox.showinfo("Kurs Kaydı", sonuc)

def main():
    root = tk.Tk()
    app = EgitimGUI(root)
    root.mainloop()

if __name__ == "__main__":
    main()

```
