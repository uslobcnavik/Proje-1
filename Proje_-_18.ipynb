```python
import tkinter as tk
from tkinter import ttk, messagebox

class TarihiSahsiyet:
    def __init__(self, ad, omru, donem):
        self.ad = ad
        self.omru = omru
        self.donem = donem

class Donem:
    def __init__(self, ad, baslangic, bitis):
        self.ad = ad
        self.baslangic = baslangic
        self.bitis = bitis

class TarihOlayi:
    def __init__(self, ad, tarih, aciklama):
        self.ad = ad
        self.tarih = tarih
        self.aciklama = aciklama

class TarihciApp(tk.Tk):
    def __init__(self):
        super().__init__()
        self.title('Tarihçi - Tarihi Olaylar Veritabanı')
        self.geometry('1000x700')
        self.configure(bg='#333')
        self.sahsiyetler = []
        self.donemler = []
        self.olaylar = []
        self.verileri_yukle()
        self.arayuz_olustur()

    def verileri_yukle(self):
        self.sahsiyetler.append(TarihiSahsiyet("Jül Sezar", "M.Ö. 100 - M.Ö. 44", "Roma Cumhuriyeti'nden Roma İmparatorluğu'na"))
        self.sahsiyetler.append(TarihiSahsiyet("Attila", "406 - 453", "Geç Antik Çağ"))
        self.donemler.append(Donem("Roma İmparatorluğu", "M.Ö. 27", "M.S. 476"))

    def arayuz_olustur(self):
        self.sekme_kontrol = ttk.Notebook(self)
        self.olay_sekmesi = ttk.Frame(self.sekme_kontrol)
        self.sahsiyet_sekmesi = ttk.Frame(self.sekme_kontrol)
        self.donem_sekmesi = ttk.Frame(self.sekme_kontrol)

        self.sekme_kontrol.add(self.olay_sekmesi, text='Olaylar')
        self.sekme_kontrol.add(self.sahsiyet_sekmesi, text='Şahsiyetler')
        self.sekme_kontrol.add(self.donem_sekmesi, text='Dönemler')
        self.sekme_kontrol.pack(expand=1, fill='both')

        self.sahsiyet_listesi = ttk.Treeview(self.sahsiyet_sekmesi, columns=('Ad', 'Ömrü', 'Dönem'), show='headings')
        self.sahsiyet_listesi.heading('Ad', text='Ad')
        self.sahsiyet_listesi.heading('Ömrü', text='Ömrü')
        self.sahsiyet_listesi.heading('Dönem', text='Dönem')
        self.sahsiyet_listesi.pack(expand=True, fill='both')

        self.donem_listesi = ttk.Treeview(self.donem_sekmesi, columns=('Ad', 'Başlangıç', 'Bitiş'), show='headings')
        self.donem_listesi.heading('Ad', text='Ad')
        self.donem_listesi.heading('Başlangıç', text='Başlangıç')
        self.donem_listesi.heading('Bitiş', text='Bitiş')
        self.donem_listesi.pack(expand=True, fill='both')

        for sahsiyet in self.sahsiyetler:
            self.sahsiyet_listesi.insert('', tk.END, values=(sahsiyet.ad, sahsiyet.omru, sahsiyet.donem))
        for donem in self.donemler:
            self.donem_listesi.insert('', tk.END, values=(donem.ad, donem.baslangic, donem.bitis))

        ttk.Label(self.olay_sekmesi, text='Olay Adı:').pack(pady=10)
        self.olay_adi_girdisi = ttk.Entry(self.olay_sekmesi)
        self.olay_adi_girdisi.pack(pady=10)
        self.olay_ekle_butonu = ttk.Button(self.olay_sekmesi, text='Olay Ekle', command=self.olay_ekle)
        self.olay_ekle_butonu.pack(pady=20)

    def olay_ekle(self):
        ad = self.olay_adi_girdisi.get()
        if ad:
            yeni_olay = TarihOlayi(ad, "Belirtilmedi", "Açıklama yok")
            self.olaylar.append(yeni_olay)
            messagebox.showinfo('Başarılı', f'{ad} başarıyla eklendi.')
            self.olay_adi_girdisi.delete(0, tk.END)
        else:
            messagebox.showerror('Hata', 'Olay adı boş bırakılamaz.')

if __name__ == "__main__":
    app = TarihciApp()
    app.mainloop()
```
